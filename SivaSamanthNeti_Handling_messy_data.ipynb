# TASK 1: Clean the Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])
df = df.drop_duplicates()

print(f"Cleaned Data: \n{df}")
```

# TASK 2: Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.get_dummies(df, columns = ['city'])

le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])

print(f"Encoded Data: {df}")
```

# TASK 3: Scale Numerical Features

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
cat_cols = ['age', 'annual_income']

minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[cat_cols] = minmax.fit_transform(df[cat_cols])

print(f"Min-Max Scaled: \n{df_minmax}")
```
